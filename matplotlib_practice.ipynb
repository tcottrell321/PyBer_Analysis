```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot (asumes cells 1 and 2 were run to import matplotlib)
# Make data green and define the legend label as Boston
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Display the legend.
plt.legend()
```

```python
# Create a Horizontal Bar Chart Using MATLIBPLOT METHOD
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
# Invert the y_axis months order with Jan at top rather than Dec
# The plt.gca means "get current axis"
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.ylabel("Month")
plt.xlabel("Fare ($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Do same thing for HORIZONTAL BAR CHART USING OBJECT ORIENTED METHOD
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')

# This line displays the legend defined in line above as New York
ax.legend()
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title("ByPer Fare by Month")
ax.invert_yaxis()

# What is the equivalent command for OOP method to reverse months on plot
# In MatLAB method its # The plt.gca means "get current axis"
# plt.gca().invert_yaxis()
```

```python
# SCATTER PLOTS using MATPLOTLIB METHOD
# Two methods exist within this Method - that is plt.plot OR  plt.scatter

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# If we don't put in the "o" parameter, this command defaults to a line chart
# plt.plot(x_axis, y_axis, 'o')

# This code produces the same thing - a scatter plot as above. 
# plt.scatter(x_axis, y_axis)
# Fill code with formatting
plt.scatter(y_axis, x_axis, color="red", label="Chicago")

#Additional Chart Formatting Code
# The plt.gca means "get current axis"
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare ($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# BUBBLE PLOTS using MATPLOTLIB METHOD
# matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, 
# vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, *, 
# plotnonfinite=False, data=None, **kwargs)[source]

# plt.scatter(x_axis, y_axis, s=y_axis_larger)

# Making data points larger with code below
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)

# Alternate way to enlarge the bubbles
#    new_list = [expression for item in list if conditional]
# To make the new_list in actual code would look like the following: 
#    plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# SCATTER AND BUBBLE PLOTS USING OOP METHOD
# fig, ax = plt.subplots()
# ax.scatter(x_axis, y_axis)

# BUBBLE PLOT - Add S parameter just like in MATPLOTLIB method
fig, ax = plt.subplots()

# Making data points larger with code below
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*5)

ax.scatter(y_axis, x_axis, color='blue', label="Boston", s=y_axis_larger)

# This line displays the legend defined in line above as New York
ax.legend()
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title("ByPer Fare by Month")
ax.invert_yaxis()
ax.set_xlim(0, 50)
```
